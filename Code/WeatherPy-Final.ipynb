```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create a test API to not exceed websites retreival limits
test = f"http://api.openweathermap.org/data/2.5/weather?q=port%20alfred&APPID={api_key}&units=imperial"
test_response = requests.get(test).json()
test_temp = test_response
test_temp
```

```python
#decalre end point and city number counter
url = "http://api.openweathermap.org/data/2.5/weather?&units=imperial&q="
cityNum = 1

#Decalare desired variables that we want to retreive from the weather website
country = []
cityy = []
date = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
latitude = []
longtitude = []



#Get rid of spaces and replace with %20
unspacedCities = []

for city in cities:
    word = []
    for letter in city:
        if letter == ' ':
            word.append('%20')
        else:
            word.append(letter)
    word = ''.join(word)
    unspacedCities.append(word)

    
print("Beginning of Data Retrieval")
print("-----------------------------")    
    
    
    
#Call API
for city in unspacedCities:
    print(f"Processing city number {cityNum}:{city}")
    cityNum += 1
    
    try:
        response = requests.get(f"{url}{city}&APPID={api_key}").json()

        temp =         response["main"]["temp"]
        humid =        response["main"]["humidity"]
        cloud =        response["clouds"]["all"]
        wind_sp =      response["wind"]['speed']
        country_name = response["sys"]["country"]
        city_name =    response["name"]
        datee =        response["dt"]
        lat =          response["coord"]["lat"]
        lng =          response["coord"]["lon"]

        temperature.append(temp) 
        humidity.append(humid)
        cloudiness.append(cloud)
        wind_speed.append(wind_sp)
        latitude.append(lat)
        longtitude.append(lng)
        country.append(country_name)
        cityy.append(city_name) 
        date.append(datee) 
        
        print("-----------------------------------") 
        time.sleep(1.01)
        
        
        #for testing, try with 50 cities instead of performing the whole loop over 580 cities
        #if cityNum >= 50:
            #break
    
    except:
        print("-------------------------------------")
        print("Record not found. Moving to next city...")
        print("Skiping........")
        print("-------------------------------------")
    continue
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cities_dic = {
    'country' : country,
    'City' : cityy,
    'Date' : date,
    'temperature' : temperature,
    'humidity' : humidity,
    'cloudiness' : cloudiness,
    'wind_speed' : wind_speed,
    'latitude' : latitude,
    'longtitude' : longtitude
}


cities_df = pd.DataFrame(cities_dic)
cities_df.head()
```
